```python
import csv

filename = 'sitka_weather_07-2021_simple.csv'
with open(filename) as f:
  reader = csv.reader(f)
  #get header row
  header_row = next(reader)
  print(header_row)

  # get column names by temperatures
  for index, column_header in enumerate(header_row):
    print(index, column_header)

  #get max temperatures
  max_tempts = []
  for row in reader:
    max_tempt = int(row[4])
    max_tempts.append(max_tempt)

print(max_tempts)

# Visualize data
import matplotlib.pyplot as plt

#Max temperatures by day in Jun
fig, ax = plt.subplots()
ax.plot(max_tempts, c='red')

plt.show()

#----- Add dates and low temperatures
from datetime import datetime
dates, max_tempts, low_tempts = [], [], []

with open(filename) as f:
  reader = csv.reader(f)
  header_row = next(reader)

  for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    max_tempt = int(row[4])
    low_tempt = int(row[5])
    dates.append(current_date)
    max_tempts.append(max_tempt)
    low_tempts.append(low_tempt)

  
fig, ax = plt.subplots()
ax.plot(dates, max_tempts, c='red')
ax.plot(dates, low_tempts, c='blue')
#add shadow between
plt.fill_between(dates, max_tempts, low_tempts, facecolor='blue', alpha=0.1)

plt.title("Daily high and low temperstures")

plt.show()



    

  
```
